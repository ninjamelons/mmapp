```python
import numpy as np
import pandas as pd

import plotly.graph_objects as go
```

```python
import spectrohandler.dataHandler as dh
```

```python
df = dh.aggregateAcquistion('title')
df.head()
```

```python
dims = df['x'].max() * 2 + 1
xy_shape = (dims, dims)
xy_shape
```

```python
nb_frames = df.iloc[:,2:].shape[1]
nb_frames
```

```python
dfmulti = df.set_index(['x','y'])
dfmulti
```

```python
dfmultire = dfmulti.values.reshape(len(dfmulti.index.levels[0]),len(dfmulti.index.levels[1]),-1).swapaxes(1,2)
dfmultire
```

```python
z_arr = list()
for i in range(nb_frames):
    arr_at = list()
    for j, multi in enumerate(dfmultire):
        arr_at.append(multi[i])
    z_arr.append(arr_at)
z_arr = np.array(z_arr)
z_arr
```

```python
len(z_arr)
```

```python
freqs = pd.melt(df.iloc[::xy_shape[0]*xy_shape[1],2:])['frequency'].values
frames = list()
for nb in range(nb_frames):
    frame = go.Frame(
        data=go.Surface(
            z=freqs[nb] * np.ones(xy_shape),
            surfacecolor=np.flipud(z_arr[nb]),
            cmin=0,
            cmax=df.iloc[:,2:].max(axis=1).max(),
            colorscale='oxy'
        ),
        name=str(freqs[nb]))
    frames.append(frame)
fig = go.Figure(frames=frames)
```

```python
#Pre-animation data
fig.add_trace(go.Surface(
    z=freqs.max() * np.ones(xy_shape),
    surfacecolor=np.flipud(z_arr[40]),
    colorscale='oxy',
    cmin=0, cmax=df.iloc[:,2:].max(axis=1).max(),
    colorbar=dict(thickness=20, ticklen=50)
    ))

def frame_args(duration):
    return {
            "frame": {"duration": duration},
            "mode": "immediate",
            "fromcurrent": True,
            "transition": {"duration": duration, "easing": "linear"},
        }


sliders = [
            {
                "pad": {"b": 10, "t": 60},
                "len": 0.9,
                "x": 1,
                "y": 0,
                "steps": [
                    {
                        "args": [[f.name], frame_args(0)],
                        "label": str(k),
                        "method": "animate",
                    }
                    for k, f in enumerate(fig.frames)
                ],
            }
        ]
```

```python
# Layout
fig.update_layout(
         title='Slices in volumetric data',
         width=800,
         height=600,
         scene=dict(
                    zaxis=dict(range=[freqs.min()-1, freqs.max()+1], autorange=False),
                    aspectratio=dict(x=1, y=1, z=1),
                    ),
         updatemenus = [
            {
                "buttons": [
                    {
                        "args": [None, frame_args(50)],
                        "label": "&#9654;", # play symbol
                        "method": "animate",
                    },
                    {
                        "args": [[None], frame_args(0)],
                        "label": "&#9724;", # pause symbol
                        "method": "animate",
                    },
                ],
                "direction": "left",
                "pad": {"r": 10, "t": 70},
                "type": "buttons",
                "x": 1,
                "y": 0,
            }
         ],
         sliders=sliders
)
fig.show()
```
